```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G= nx.Graph()
G.add_nodes_from([1,2,3,4,5])
G.add_edges_from([(1,1),(1,2), (1,3), (2,3),(3,4),(4,5),(1,5)])
nx.draw(G, with_labels =True, font_weight ='bold')
plt.show()
```

# basic graph info


```python
print("Nodes:", G.nodes)
print("Edges:", G.edges)
print("Number of Nodes:", G.number_of_nodes())
print("Number of edges:", G.number_of_edges())
```

# check if the graph is connected

```python
#check if the graph is connected
print(nx.is_connected(G))
#shortage pathe
shortest_path=nx.shortest_path(G,source=1, target=4)
print("Shortest path",shortest_path)
```

# degree of each node


```python
#degree of each node
degree=G.degree()
print("degree of each node",degree)
deg_cen=nx.degree_centrality(G)
print("degree_centrality", deg_cen)
```

```python
clustering_cofficient=nx.clustering(G)
print("clustering_cofficient",clustering_cofficient)
```

# Sub Graph

```python
sub=[1,2,3]
sub_graph=G.subgraph(sub)
nx.draw(sub_graph,with_labels =True, font_weight ='bold')
plt.show()
```
